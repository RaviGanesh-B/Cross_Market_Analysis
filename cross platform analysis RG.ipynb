```python
import requests
import pandas as pd
import time

all_records = []

for page in range(1, 6):

    url = (f"https://api.coingecko.com/api/v3/coins/markets"
           f"?vs_currency=inr&per_page=250&order=market_cap_desc"
           f"&page={page}&sparkline=False")

    response = requests.get(url)
    all_data = response.json()

    if response.status_code != 200:
       print(f"Error on page {page}: {response.status_code}")
       time.sleep(20)
       continue

    for i in all_data:
        all_records.append(dict(
            id=i["id"],
            symbol=i["symbol"],
            name=i["name"],
            current_price=i["current_price"],
            market_cap=i["market_cap"],
            market_rank=i["market_cap_rank"],
            total_volume=i["total_volume"],
            circulating_supply=i["circulating_supply"],
            total_supply=i["total_supply"],
            ath=i["ath"],
            atl=i["atl"],
            last_updated=i["last_updated"]
        ))


    time.sleep(20)


coins_df = pd.DataFrame(all_records)

coins_df["last_updated"] = pd.to_datetime(coins_df["last_updated"], errors="coerce").dt.date

coins_df

#completed all 5 pages readed
```

```python
import pandas as pd 
import requests
import time

coin_ids = ["bitcoin","ethereum","tether","binancecoin","ripple"]

all_data =[]

for coin_id in coin_ids:
    url = f"https://api.coingecko.com/api/v3/coins/{coin_id}/market_chart?vs_currency=usd&days=365"
    response = requests.get(url)
    
    if response.status_code != 200:
       print(f"Error fetching {coin_id}: {response.status_code}")
       continue
        
    data = response.json()
    
    if 'prices' not in data:
        print(f"No price data for {coin_id}: {data}")
        continue
         
    all_data.extend([{
    "coin_id": coin_id,
    "date": pd.to_datetime(ts, unit='ms').date(),
    "price_usd": price
     } 
     for ts, price in data['prices']])


    time.sleep(10)

top_df = pd.DataFrame(all_data)

top_df



#completed top 5 crypto  historical prices
```

```python
import pandas as pd

oil_spike_vs_bitcoinchange = """
                             SELECT *
                             FROM(
                             SELECT oil.date,
                             oil.price_usd AS oil_price,
                             oil.price_usd - LAG(oil.price_usd) OVER (ORDER BY oil.date) AS oil_spike,
                             bitcoin.price_usd AS bitcoin_price,
                             bitcoin.price_usd - LAG(bitcoin.price_usd) OVER (ORDER BY bitcoin.date) AS bitcoin_change
                             FROM oil_prices oil
                             JOIN crypto_prices bitcoin
                             ON oil.date = bitcoin.date
                             WHERE bitcoin.coin_id = 'bitcoin'
                             AND YEAR(oil.date) = 2025 
                             ) AS t
                             WHERE oil_spike > oil_price * 0.05 
                             ORDER BY date;
                             """

cpa_cursor.execute(oil_spike_vs_bitcoinchange)

oil_spike_vs_bitcoinchange = cpa_cursor.fetchall()

oil_price_spike_vs_bitcoin_price_change = pd.DataFrame(oil_spike_vs_bitcoinchange , columns = [col[0] for col in cpa_cursor.description])

oil_price_spike_vs_bitcoin_price_change

#Find days when oil price spiked and compare with Bitcoin price change. completed
```

```python
import pandas as pd

gspc_vs_oil = """
              SELECT
                  stock.date,
                  stock.close AS sp500_close,
                  oil.price_usd AS oil_price
              FROM stock_prices stock
              JOIN oil_prices oil
                 ON stock.date = oil.date
              WHERE stock.ticker = '^GSPC'
                AND YEAR(stock.date) = 2025
              ORDER BY stock.date;
              """
cpa_cursor.execute(gspc_vs_oil)

gspc_vs_oil = cpa_cursor.fetchall()

gspc_prices_vs_oil_prices_on_samedates = pd.DataFrame(gspc_vs_oil, columns = [col[0] for col in cpa_cursor.description])

gspc_prices_vs_oil_prices_on_samedates

#Compare stock prices (^GSPC) with crude oil prices on the same dates completed
```

```python
import pandas as pd

stock_oil_bitcoin = """
                    SELECT bitcoin.date,
                           bitcoin.price_usd AS bitcoin_price,
                           oil.price_usd AS oil_price,
                           nsei.close AS nifty_close,
                           gspc.close AS sp500_close,
                           ixic.close AS nasdaq_close
                    FROM crypto_prices bitcoin
                    JOIN oil_prices oil
                        ON bitcoin.date = oil.date
                    JOIN stock_prices nsei
                        ON bitcoin.date = nsei.date AND nsei.ticker = '^NSEI'
                    JOIN stock_prices gspc
                        ON bitcoin.date = gspc.date AND gspc.ticker = '^GSPC'
                    JOIN stock_prices ixic
                        ON bitcoin.date = ixic.date AND ixic.ticker = '^IXIC'
                    WHERE bitcoin.coin_id = 'bitcoin'
                      AND YEAR(bitcoin.date) = 2025
                    ORDER BY bitcoin.date;
                    """

cpa_cursor.execute(stock_oil_bitcoin)

stock_oil_bitcoin = cpa_cursor.fetchall()

stock_oil_bitcoin_daily_comparison = pd.DataFrame(stock_oil_bitcoin , columns = [col[0] for col in cpa_cursor.description])

stock_oil_bitcoin_daily_comparison

#Multi-join: stock prices, oil prices, and Bitcoin prices for daily comparison completed
```
